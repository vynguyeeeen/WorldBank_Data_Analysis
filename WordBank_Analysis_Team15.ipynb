## 	Strategy for Missing Values & Outliers
To gain a greater understanding of the Middle East, we investigated the World Bank’s dataset, which originally came with various missing values. To salvage the issue, we opted to fill in missing values using the median, the ‘norm’, of each region. We chose this method since the missing values will cause the data distribution to be skewed and contain outliers. Statistically, when there is a skewed distribution, the median is a better measure of central tendency than the mean  (Frost, 2020).We considered manually entering data points discovered from external resources, but this would significantly disrupt the current dataset as the sources itself, collection methods, and potential biases surrounding those data points would be in question. To safely move forward, we checked if the altered dataset would drastically impact the original data. To our favor, it had little impact. Although we were only able to fill in 93% of missing values, we discovered that certain features, such as literacy rate, had minimal data points to begin with. Thus, it was impossible to calculate the norm.  

Diving deeper, both our regions and the world shared similar potential outliers. If all regions had similar results, this meant that the data collected was scattered, thus, these ‘potential outliers’ are in fact normal. Nonetheless, there was a potential outlier in the percentage of people who received antiretroviral therapy coverage (ATC) that noticeably played a role in our region but had minimal effect on the rest of the world. We discovered that Romania was the only high-income country in the region to provide universal access to HIV treatment. Romania saw an increase in long-term survival rate in HIV infected persons who received ATC and showcased a massive improvement among HIV patients. The number of patients per therapy increased from 3% in 1995 to 60% in 2001 (Frost, 2020); such findings matched with our dataset. 
 
##   Country that best represents the region
Before finding the best country that represented the region, we defined a benchmark that will be used as the central point of our analysis. The benchmark is the observation aggregated by a descriptive statistic of the numerical features of the 14 countries that constitute the region. From our previous missing values and outlier analysis, we saw that the missing values imputation had little effect on the mean and median statistic of the distribution. 
To control for outliers, we decided to use the median because of its more robust attribute when dealing with outliers to build our synthetic benchmark. With this, we were able to choose the “best country” based on the hypothesis that the country with the features that are closest to benchmark will be the country that best approximates our region. Prior to the analysis, we standardized the dataset to have each feature in the same scale and ready to compare all the features cross-sectionally. 

For this, we used the Z-score formula, where we used the mean and standard deviation of each distribution to compute the Z-score for each value. With the standardized dataset, we then computed an array of “weights” that distributed equally for each of the numerical features to analyze (out of 41 numerical features, we had an array that had values of 1/41 and its total summed to 1) this way we assumed all columns were being considered the same when aggregating the difference between a country’s feature value with that of the benchmark. We then computed the squared difference of each of the values against that of the benchmark and did a horizontal weighted sum to get the total amount of cross-sectional squared error for each country. 

We then took the square root of the total squared error to get a sorted list of the countries with the least amount of total root squared error. As a result, Uzbekistan became the country with the least amount of total root squared error, categorized statistically as the country that is the closest approximation to the benchmark that we created. Thus, Uzbekistan best represents “on average” our region.

### Flag Potential Outlier(s) in Features

```python
# Automating Outlier detection and labeling

def out_detection(df,boundary_factor=1.5):

    '''
    Inter-Quartile Range (IQR) Method for Outlier detection for quantitative variables:
    
    Creates a theoretical range of lower and upper boundaries by assuming that every data 
    that is outside 2.7 standard deviations from the mean of the distribution is an outlier. 
    
    Therefore:
    
    Lower boundary: Q1 - (1.5 * IQR)
    Upper boundary: Q3 + (1.5 * IQR)
    
    Where:
        IQR = Q3 - Q1
    
    '''

    # Making a copy of df 
    df_with_outs = df.copy()

    # Initializing dataframe to store columns with number of outliers, if any
    out_df = pd.DataFrame()

    # Iterating over columns of dataframe
    for col in df.columns:
        
        # Making sure column is of type int or float
        if ((df_with_outs[col].dtypes == 'float64')|(df_with_outs[col].dtypes == 'int64')):
        
            # Calculating value for first quantile of column's distribution
            q1 = df_with_outs[col].describe().loc["25%"] 

            # Calculating value for third quantile of column's distribution
            q3 = df_with_outs[col].describe().loc["75%"] 

            # Calculating IQR
            iqr = q3-q1

            # Calculating left threshold for outliers
            left_out_threshold = q1 - (iqr * boundary_factor)

            # Calculating right threshold for outliers
            right_out_threshold = q3 + (iqr * boundary_factor)

            # Creating new column that labels which row is an outlier for the column being analyzed

            # if values in columns are outside of the left or right threshold, label them as outliers
            out_col = ((df_with_outs[col] < left_out_threshold) | (df_with_outs[col] > right_out_threshold)).astype(int)

            # Sum of all outlier values, if any
            out_sum = out_col.sum()

            if out_sum > 0:
                df_with_outs['out_{}'.format(col)] = out_col
                
                out_df.loc[col,'Outliers'] = out_sum # adding column with number of outliers
                out_df.loc[col,'Total'] = df[col].shape[0] # getting total number of rows
                out_df.loc[col,'Outliers Percentage'] = (out_df.loc[col,'Outliers'] / out_df.loc[col,'Total']).round(2)
                out_df.loc[col, 'left_out_threshold'] = left_out_threshold
                out_df.loc[col, 'right_out_threshold'] = right_out_threshold
    
    return df_with_outs, out_df
```

```python
# Calling function to receive a dataframe with outliers count and labels
df_with_outs, out_df = out_detection(df=mirage.iloc[:,:44])
```

## Detect Best Country 'on Average' 

```python
# Storing the region benchmark name 
benchmark = 'Middle East & Russia'

# choosing quantitative values only
median_df = mirage.iloc[:,:44].iloc[:,3:]
```

```python
# Creating an aggregate row called "Middle East & Russia" as an aggregate 
# of the median of all the countries that constitute the region
# this row will act as our benchmark to choose which country approximates the most to this
median_df.loc[benchmark] = median_df.median()
median_df.loc[[benchmark]].round(2)
```

```python
# standardizing the dataframe into Z-scores to make sure all columns can be compared cross-sectionally
standard_median_df = (median_df - median_df.mean())/median_df.std()
standard_median_df.head(5).round(2)
```

```python
# create feature weight array (assumes there is equal weight for all features)
feature_weight = np.ones(standard_median_df.shape[1])*(1/standard_median_df.shape[1])

# get difference between all countries and the median standard values from the aggregate region row
# this helps us compare the standardized values of the benchmark against all the countries
# those countries with the least aggregate difference between the benchmark values approximates the most statistically
benchmark_difference = (standard_median_df - standard_median_df.loc[benchmark])
weighted_sum = ((benchmark_difference**2)*feature_weight).sum(axis=1)
 
# calculating the root square error
root_square_error = np.power(weighted_sum,1/2).sort_values()
```

```python
# Adding Country Name column to the root_square dataframe
root_square_df = pd.DataFrame(root_square_error,columns=['Root Square Error'])
root_square_df['Country Name'] = mirage['Country Name']

# Showing top 5 countriest with the least amount of root square error
root_square_df.head(5).round(2)
```

Uzbekistan appears to be the country with the least amount of total root square error against the benchmark we created for the Middle East and Russia. Therefore, we conclude Uzbekistan is the country that best represents "on average" the Middle East and Russia.
